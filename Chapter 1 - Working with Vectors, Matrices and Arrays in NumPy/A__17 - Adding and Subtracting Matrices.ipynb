# A17: Adding and Subtracting Matrices

```python
# Problem Statement: To add or subtract two matrices
# Solution: Use numPy's add and subtract
```

```python
# Load the library

import numpy as np
```

```python
# create the matrix A

matrix_A = np.array([[1, 1, 1],
                     [1, 1, 1],
                     [1, 1, 2]])
```

```python
# create matrix B

matrix_B = np.array([[1, 3, 1], 
                    [1, 3, 1],
                    [1, 3, 8]])
```

```python
# Add two matrices

matrix_Add = np.add(matrix_A, matrix_B)
```

```python
# print the result of adding two matrices

print(matrix_Add)
```

```python
# Subtract two matrices

matrix_Sub =np.subtract(matrix_A, matrix_B)
```

```python
# print the result of subtracting two matrices

print(matrix_Sub)
```

```python
# + and - operators can also be used
```

```python
matrix_A + matrix_B
```

```python
matrix_A - matrix_B
```
