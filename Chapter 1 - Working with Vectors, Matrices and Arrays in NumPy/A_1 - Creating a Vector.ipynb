# A1: Creating a Vector

```python
# Vector is an array with single dimension
```

```python
# Problem Statement : To create a vector
# Solution : Use NumPy to create a one-dimensional array
```

```python
# Loading the library

import numpy as np
```

```python
# create a vector as a row

vec_row = np.array([1, 2, 3])
```

```python
# print the vector row

print(vec_row)
```

```python
# create a vector as a column

vec_col = np.array([[1],
                  [2],
                  [3],
                  [4]])
```

```python
# print the vector column

print(vec_col)
```
