# A14: Getting the Diagonal of a Matrix

```python
# Problem Statement: to get the diagonal elements of a matrix
# Solution: use diagonal
```

```python
# load the library

import numpy as np
```

```python
# create the matrix

matrix = np.array([[1, 2, 3], 
                  [4, 5, 6],
                  [6, 7, 8]])
```

```python
# print the matrix

print(matrix)
```

```python
# reurn the diagonal elements

matrix.diagonal()
```

```python
# possible to diagonal off from the main diagonal using the offset parameter
```

```python
# return the diagonal one above the main diagonal

matrix.diagonal(offset = 1)
```

```python
# return the diagonal one below the main diagonal

matrix.diagonal(offset = -1)
```
