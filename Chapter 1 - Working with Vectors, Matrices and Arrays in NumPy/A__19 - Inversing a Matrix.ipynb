# A19: Inversing a Matrix

```python
# Problem Statement: To calculate the inverse of a square matrix
# Solution: Use numPy's linear algebra inv method
```

```python
# Load the library

import numpy as np
```

```python
# create the matrix

matrix_A = np.array([[1, 4], 
                    [2, 5]])
```

```python
# calculate the inverse of a matrix

np.linalg.inv(matrix_A)
```

```python
# multiply matrix and its inverse
# result is a identity matrix

matrix_A @ np.linalg.inv(matrix_A)
```
