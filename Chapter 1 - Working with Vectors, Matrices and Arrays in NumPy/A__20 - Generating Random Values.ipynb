# A20: Generating Random Values

```python
# Probelm Statement: to generate pseudorandom values
# Solution: Use numPy's random
```

```python
# Load the library

import numpy as np
```

```python
# Set seed

np.random.seed(0)
```

```python
# generate three random floats beteen 0.0 and 1.0

np.random.random(3)
```

```python
# to generate integers between 0 and 10

np.random.randint([5, 26, 12])
```

```python
# generate numbers from a distribution
```

```python
# generate three numbers from a normal distribution with mean 0.1, and standard deviation of 1.0

np.random.normal(0.1, 1.0, 3)
```

```python
# generate four numbers from a logistic distribution with mean 0.0 and scale of 1.0

np.random.logistic(0.0, 1.0, 4)
```

```python
# generate three numbers greater than or equal to 1.0 and less than 2.0

np.random.uniform(1.0, 2.0, 3)
```
