# A12: Flattening a Matrix

```python
# Problem Statement: to transform a matrix into a one-dimensional array
# Solution: Use flatten
```

```python
# Load the library

import numpy as np
```

```python
# create the matrix

matrix = np.array([[1, 2, 3],
                  [5, 6, 7],
                  [6, 7, 8]])
```

```python
# flatten the matrix

matrix.flatten()
```

```python
# alternatively, reshape to create a row vector

matrix.reshape(1, -1)
```

```python
# method 2 - ravel returns a copy of the original array
# ravel operates on the original object itself 
# flatten of arrays
```

```python
# create matrix A

matrix_A = np.array([[1, 2],
                    [3, 4]])
```

```python
# print matrix A

print(matrix_A)
```

```python
# create matrix B

matrix_B = np.array([[5, 6],
                    [7, 8]])
```

```python
# print matrix B

print(matrix_B)
```

```python
# create a list of matrices

matrix_list = [matrix_A, matrix_A]
```

```python
# print the matrix list

print(matrix_list)
```

```python
# flatten the entire list of matrices

np.ravel(matrix_list)
```
