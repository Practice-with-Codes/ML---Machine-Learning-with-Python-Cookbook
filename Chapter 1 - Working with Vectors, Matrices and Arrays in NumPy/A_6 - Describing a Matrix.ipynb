# A6: Describing a Matrix

```python
# Problem Statement: to describe the shape, size and dimension of the marix
# Solution: Use shape, size and ndim attributes of numPy
```

```python
# Load the library

import numpy as np
```

```python
# create the matrix

matrix = np.array([[1, 2, 3, 4],
                  [4, 5, 6, 7],
                  [7, 8, 9, 10]])
```

```python
# print the matrix

print(matrix)
```

```python
# view the number of rows and columns

matrix.shape
```

```python
# view the number of elements (rows * columns)

matrix.size
```

```python
# view the number of dimensions

matrix.ndim
```
