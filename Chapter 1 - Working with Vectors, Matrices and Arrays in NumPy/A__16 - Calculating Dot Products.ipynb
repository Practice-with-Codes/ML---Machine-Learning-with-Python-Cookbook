# A16: Calculating Dot Products

```python
# Problem Statement: to calculate the dot product of two vectors
# Solution: Use numPy's dot
```

```python
# Load the library

import numpy as np
```

```python
# create vector A

vector_A = np.array([1, 2, 3])
```

```python
# print vector A

print(vector_A)
```

```python
# create vector B

vector_B = np.array([5, 6, 7])
```

```python
# print vector B

print(vector_B)
```

```python
# calculate the dot product

np.dot(vector_A, vector_B)
```

```python
# alternatively, calculate the dot product using @ operator

vector_A @ vector_B
```
