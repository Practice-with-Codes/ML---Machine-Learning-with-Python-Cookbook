# A18: Multiplying Matrices

```python
# Problem Statement: To multiply two matrices
# Solution: Use numPy's dot
```

```python
# Load the library

import numpy as np
```

```python
# create matrix A

matrix_A = np.array([[1, 2], 
                    [1, 3]])
```

```python
# print matrix A

print(matrix_A)
```

```python
# create matrix B

matrix_B = np.array([[1, 2], 
                    [1, 5]])
```

```python
# print matrix B

print(matrix_B)
```

```python
# multiply the matrices

matrix_C = np.dot(matrix_A, matrix_B)
```

```python
# print the result, matrix C

print(matrix_C)
```

```python
# multiply two matrices using @ operator

matrix_A @ matrix_B
```

```python
# multiply two matrices element-wise using * operator

matrix_A * matrix_B
```
