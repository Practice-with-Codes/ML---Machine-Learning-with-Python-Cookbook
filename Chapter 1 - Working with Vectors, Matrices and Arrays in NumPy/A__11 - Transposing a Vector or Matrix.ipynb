# A11: Transposing a Vector or Matrix

```python
# Problem Statement: to transpose a vector or a matrix
# Solution: Use T method
```

```python
# Load the library

import numpy as np
```

```python
# create a matrix

matrix = np.array([[1, 2, 3], 
                  [4, 5, 6], 
                  [6, 7, 8]])
```

```python
# transpose of a matrix

matrix.T
```

```python
# vector cannot be transposed as it is a collection of values
```

```python
# Transpose vector

np.array([1, 2, 3, 4, 5, 6]).T
```

```python
# Transposing a vector --> converting  a row vector to a column vector
```

```python
# Transpose row vector

np.array([[1, 2, 3, 4, 5, 6]]).T
```
