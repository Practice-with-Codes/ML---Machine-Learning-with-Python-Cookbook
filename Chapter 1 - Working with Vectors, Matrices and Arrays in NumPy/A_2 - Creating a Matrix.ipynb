# A2: Creating a Matrix

```python
# Problem Statement : To create a matrix
# Solution: Use NumPy to create a two-dimensional array
```

```python
# Loading the library

import numpy as np
```

```python
# Create a matrix

matrix_a = np.array([[1, 2], 
                    [1, 2], 
                    [1, 2]])
```

```python
# print the matrix a

print(matrix_a)
```

```python
# matrix data structure
# Note: Majority of the NumPy operations return array, not matrix objects

matrix_object = np.mat([[1, 2], 
                       [1, 2],
                       [1, 2]])
```

```python
# print the matrix object

print(matrix_object)
```
