# A4: Pre-allocating Numpy Arrays

```python
# Problem Statement: to pre-allocate arrays of a given size with some value
# Solution: numPy has functions for generating vctors and matrices of any size using 0's and 1's 
```

```python
# Load the library

import numpy as np
```

```python
# generate a vector of shape(1, 5) containing all zeroes

vector = np.zeros(shape = 5)
```

```python
# print the vector

print(vector)
```

```python
# generate a matrix of shape(3, 3) containing all ones

matrix = np.full(shape = (3, 3), fill_value = 1)
```

```python
# print the vector\

print(matrix)
```
