# A3: Creating a Sparse Matrix

```python
# Problem Statement: Given data with non-zero values, efficiently represent it
# Solution: create a sparse matrix
```

```python
# Load the library

import numpy as np
from scipy import sparse
```

```python
# create the matrix

matrix = np.array([[0, 0],
                  [0, 1],
                  [3, 0]])
```

```python
# print the matrix

print(matrix)
```

```python
# create the compressed sparse row (CSR) matrix

mat_sparse = sparse.csr_matrix(matrix)
```

```python
# print the sparse matrix

print(mat_sparse)
```
