# AD__13: Function to Check Check Odd or Even Number

```python
def even(number):
    if number == 0:
        return True
    return odd(number - 1)
def odd(number):
    if number == 0:
        return False
    return even(number - 1)
def isEven(number):
    if even(number):
        print("Number is Even")
    else:
        print("Number is Odd")
```

```python
even(5)
```

```python
odd(9)
```

```python
isEven(2)
```

```python
isEven(7)
```
