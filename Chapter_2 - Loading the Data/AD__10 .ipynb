# AD__10: Function to Find the Armstrong Numbers in a Given Limit

```python
def Armstrong_Number():
    num = int(input("Enter the number to check for Armstrong:"))
    digit = len(str(num))
    f = num
    sum = 0
    while (f!=0):
        a = f % 10
        f = f // 10
        sum = sum + (a ** digit)
    if (sum == num):
        print(num, "is an armstrong number ")
    else:
        print(num, "is not an armstrong number")
        
Armstrong_Number()        
```
