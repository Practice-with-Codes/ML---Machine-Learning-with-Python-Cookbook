# B_9: Querying a SQLite Database

```python
# Problem Statement: To load data from a database using SQL
# Solution: Use pandas read_sql_query to make a SQl query to a database and load it
```

```python
# load the library

import pandas as pd
from sqlalchemy import create_engine
```

```python
# create a connection to the database

data_connection = create_engine('sqlite:///sample.db')
```

```python
df = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", data_connection)
```
