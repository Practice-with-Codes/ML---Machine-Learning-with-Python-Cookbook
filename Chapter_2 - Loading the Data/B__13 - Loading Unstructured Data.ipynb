# B_13: Loading Unstructured Data

```python
# Problem Statement: To load in unstructured data like text or images
# Solution: Use the base Python open function to load the information
```

```python
# import the libraries

import requests
```

```python
# url to download the text file from

text_url = 'https://machine-learning-python-cookbook.s3.amazonaws.com/text.txt'
```

```python
# get the txt file

r = requests.get(text_url)
```

```python
# write it to text.txt locally

with open('text.txt', 'wb') as f:
    f.write(r.content)
```

```python
# read in the file

with open('text.txt', 'r') as f:
    text = f.read()
```

```python
# print the content

print(text)
```
