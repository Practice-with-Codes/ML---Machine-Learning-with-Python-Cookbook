# B_12: Loading Data from an S3 Bucket

```python
# Problem Statement: To read a CSV file from an S3 bucket you have to access to
# Solution: Add storage options to pandas giving it access to the S3 object
```

```python
# import the libraries

import pandas as pd
```

```python
# s3 path to csv

s3_uri = 's3://machine-learning-python-cookbook/data.csv'
```

```python
# set AWS credentials

ACCESS_KEY_ID = "hello"
SECRET_ACCESS_KEY = "12345"
```

```python
# read the csv into a dataframe

dataframe = pd.read_csv(s3_uri, storage_options = {"key": ACCESS_KEY_ID, "secret": SECRET_ACCESS_KEY})
```
