# B_7: Loading a Avro File

```python
# Problem Statement: To load an avro file into a pandas dataframe
# Solution: Use the panda avro library read_avro method
```

```python
# load the library

from fastavro import reader
import pandas as pd
import requests
from io import BytesIO
```

```python
# create the url

url = 'https://machine-learning-python-cookbook.s3.amazonaws.com/data.avro'
```

```python
# Download AVRO file
data = requests.get(url).content
```

```python
# Read AVRO records
records = list(reader(BytesIO(data)))
```

```python
# Convert to Pandas DataFrame
df = pd.DataFrame(records)
```

```python
# view the first two rows

print(df.head())        
```
