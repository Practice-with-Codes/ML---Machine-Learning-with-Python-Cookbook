# A_4: Pre-allocating NumPy Arrays

```python
# Problem Statement: To pre-allocate arrays of given size with some value
# Solution: NumPy has functions for generating vectors and matrices of any size using 0's and 1's or any number of your choice
```

```python
# load the library

import numpy as np
```

```python
# Generate a vector of shape(1, 5) containing all zeroes

vector = np.zeros(shape = 5)
```

```python
# view the vector

print(vector)
```

```python
# Generate a vector of shape (1, 5) containing all ones

vector = np.ones(shape = 5)
```

```python
# view the vector

print(vector)
```

```python
# Generate a vector of shape (3,3) containing all ones

matrix = np.full(shape = (3, 3), fill_value = 1)
```

```python
# view the matrix

print(matrix)
```
