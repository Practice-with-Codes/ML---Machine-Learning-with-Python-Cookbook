# A_1: Creating a Vector

```python
# Problem Statement: To create a vector
# Solution: Use NumPy to create a one-dimensional array
```

```python
# Load the library

import numpy as np
```

```python
# Create a vector as a row

vector_row = np.array([1, 2, 3])
```

```python
# print the row vector

print(vector_row)
```

```python
# create a vector as a column

vector_col = np.array([[1], 
                      [2], 
                      [3]])
```

```python
# print the column vector

print(vector_col)
```

```python
# References: 
# Vector - An array with single dimension
# To create a vector, create a one-dimensional array
# Like vectors, arrays can be represented horizontally(rows) and vertically(columns)
```
