# A__19: Inverting a Matrix

```python
# Problem Statement: To calculate the inverse of a square matrix
# Solution: Use NumPy's linear algebra inv method
```

```python
# load the library

import numpy as np
```

```python
# create the matrix

matrix = np.array([[1, 4], 
                  [2, 5]])
```

```python
# view the matrix

print(matrix)
```

```python
# calculate the inverse of matrix

np.linalg.inv(matrix)
```

```python
# multiply matrix and its inverse

matrix @ np.linalg.inv(matrix)
```
