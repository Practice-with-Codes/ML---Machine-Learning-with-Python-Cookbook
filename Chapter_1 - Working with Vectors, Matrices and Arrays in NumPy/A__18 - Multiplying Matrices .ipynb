# A__18: Multiplying Matrices 

```python
# Problem Statement: To multiply two matrices
# Solution: Use NumPy's dot
```

```python
# load the library

import numpy as np
```

```python
# create the first matrix

matrix_a = np.array([[1, 1], 
                    [1, 2]])
```

```python
# view the first matrix

print(matrix_a)
```

```python
# create the second matrix

matrix_b = np.array([[3, 4], 
                    [5, 6]])
```

```python
# view the second matrix

print(matrix_b)
```

```python
# multiply two matrices

np.dot(matrix_a, matrix_b)
```

```python
# Alternatively, use the @ operator

matrix_a @ matrix_b
```

```python
# to do element-wise multiplication, use * operator

matrix_a * matrix_b
```
