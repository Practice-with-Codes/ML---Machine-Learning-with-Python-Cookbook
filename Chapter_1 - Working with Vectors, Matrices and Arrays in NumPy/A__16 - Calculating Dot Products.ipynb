# A__16: Calculating Dot Products

```python
# Problem Statement: To calculate the dot product of two vectors
# Solution: Use NumPy's dot
```

```python
# load the library

import numpy as np
```

```python
# create the first vector

vector_a = np.array([1, 2, 3])
```

```python
# view the first vector

print(vector_a)
```

```python
# create the second vector

vector_b = np.array([4, 5, 6])
```

```python
# view the second vector

print(vector_b)
```

```python
# calculate the dot product

np.dot(vector_a, vector_b)
```

```python
# Alternatively, use @ operator

vector_a @ vector_b
```
