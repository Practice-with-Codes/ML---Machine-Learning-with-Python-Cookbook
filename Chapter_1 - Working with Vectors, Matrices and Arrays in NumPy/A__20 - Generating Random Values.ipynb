# A__20: Generating Random Values

```python
# Problem Statement: To generate pseudorandom numbers
# Solution: Use NumPy's random
```

```python
# load the library

import numpy as np
```

```python
# set seed

np.random.seed(0)
```

```python
# generate three random floats between 0.0 and 1.0

np.random.random(3)
```

```python
# common to generate intergers, other than floats
```

```python
# generate three random numbers between 0 and 10

np.random.randint(0, 11, 3)
```

```python
# generate numbers from drawing them from a distribution
```

```python
# draw three numbers from a normal distribution with mean 0.0 and standard deviation 1.0

np.random.normal(0.0, 1.0, 3)
```

```python
# draw three numbers from a logistic distribution with mean 0.0 and standard deviation 1.0

np.random.logistic(0.0, 1.0, 3)
```

```python
# draw three numbers greater than or equal to 1.0 and less than 2.0

np.random.uniform(1.0, 2.0, 3)
```
