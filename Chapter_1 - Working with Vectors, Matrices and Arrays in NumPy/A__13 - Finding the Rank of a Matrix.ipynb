# A__13: Finding the Rank of a Matrix 

```python
# Problem Statement: To know the rank of a matrix
# Solution: Use NumPy's linear algebra method matrix_rank
```

```python
# load the library

import numpy as np
```

```python
# create the matrix

matrix = np.array([[1, 1, 1], 
                  [1, 1, 10], 
                  [1, 1, 15]])
```

```python
# view the matrix

print(matrix)
```

```python
# return the matrix rank

np.linalg.matrix_rank(matrix)
```

```python
# References:
# The rank of matrix is the dimensions of the vector space spanned by its columns or rows
```
